```python
import sympy
import matplotlib.pyplot as plt
import numpy as np
import os


def f(x):
    return x**4-5*x**3+7*x**2-3*x

 #representar graficamente

X=np.linspace(0,3,200)
Y=f(X)
fig, ax = plt.subplots() # Una figura con un eje
ax.set_xlabel('x')
ax.set_ylabel('y',rotation=None)
ax.plot(X,Y)
# Obtenemos una función símbolica... ¿Como?
x=sympy.symbols('x')
fp=sympy.diff(f(x),x)
fpp=sympy.diff(fp,x)
fppn=sympy.lambdify(x,fpp)
pcs=sympy.solve(fp,x)
print(f'Se han encontrado {len(pcs)} puntos críticos')
for pc in pcs:
    pcn=float(pc)
    deriv2=fppn(pcn)
    if deriv2 <0:
        tipo,marker ='maximo','bo'
    elif deriv2 >0:
        tipo, marker ='minimo', 'ro'
    else:
        tipo, marker='indeterminado','kx'
        # \u2248 es el código unicode de "aproximadamente igual"
    print(f'{tipo} en x={pc} \u2248 {float(pc)}')
    ax.plot(pcn,f(pcn),marker)

# fname,extension=(os.path.splitext(os.path.basename(__file__)))
# plt.savefig(fname+'.eps',format='eps',papertype='a6')
```
